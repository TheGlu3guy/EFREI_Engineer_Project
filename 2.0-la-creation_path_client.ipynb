Cas général

```python
# Group by token pour avoir les différents tokens liés aux utilisateurs puis nouveau group by l'id pour avoir le nombre de token différents par utilisateur
user_ntoken = df.groupby(["id_utilisateur", "token_1", "token_2"]).size().groupby("id_utilisateur").size().sort_values(ascending=False).reset_index(name="nb_token")
user_ntoken
```

```python
# Un utilisateur a beaucoup trop de token par rapport aux autres, suspect...
```

```python
#nombre de lignes par user
user_lines = df.groupby('id_utilisateur').size().sort_values(ascending=False).reset_index(name="nb_lines")
user_lines
```

```python
user_ntoken.merge(user_lines, on="id_utilisateur", how="left").sort_values(by="nb_lines", ascending=False).head(20)
```
